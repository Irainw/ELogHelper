```python
import requests
import pandas as pd
import re
```

```python
def removeObjects(given_text_as_string):
	regexObject = '<[^>]+>'
	given_text_objects_removed = re.sub(regexObject, " ", given_text_as_string, 0)
	return(given_text_objects_removed)

def removeURL(given_text_as_string):
	regexURL = "(?P<url>http[s]?://[^\s]+)"
	given_text_URL_removed = re.sub(regexURL, " ", given_text_as_string, 0)
	return(given_text_URL_removed)

def removeNewLine(given_text_as_string):
	given_text_newline_removed = given_text_as_string.replace('\n', ' ')
	return(given_text_newline_removed)
```

```python
params = {
    'after':'1',
    'limit':'10',
    'logbook':'SPEAR3',
}

r = requests.get('https://mccelog.slac.stanford.edu/elog/wbin/elog_display_json.php', params=params)
```

```python
rList = r.json()
rDF = pd.DataFrame.from_dict(rList, orient='columns')

rDF
```
